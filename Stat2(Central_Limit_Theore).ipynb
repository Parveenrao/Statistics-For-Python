```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_samples = 1000
sample_size = 30
scale =2

samples = np.random.exponential(scale=scale, size=(num_samples, sample_size))
mean_sample = np.mean(samples, axis=1)
```

```python

plt.hist(mean_sample , bins = 30 , density = True )
sns.kdeplot(mean_sample, color='red', linewidth=2)
plt.title("Central Limit Theorem - Sample Means Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
