```python
import seaborn as sns
import numpy as np
```

```python
data = sns.load_dataset('iris')
data
```

```python
sns.kdeplot(x = data['sepal_length'] , data = data)
```

```python
y_quant = []

temp = sorted(data['sepal_length'].to_list())
for i in range(1 ,100):
    y_quant.append(np.percentile(temp ,i))
```

```python
sample = np.random.normal(0 ,1 , 1000)
```

```python
x_quant = []

temp = sorted(sample)
for i in range(1 ,100):
    x_quant.append(np.percentile(temp ,i))
```

```python
sns.scatterplot(x = x_quant , y = y_quant)
```

Using Library Statsmodel.api

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create dataset with 100 values that follow a normal distribution
import statsmodels.api as sm
np.random.seed(0)
data = np.random.normal(0,1, 1000)
fig = sm.qqplot(data , line = '45')
plt.show()
plt.figure(figsize = (0,0))
```

```python
#create dataset of 100 uniformally distributed values
data = np.random.uniform(0,1, 1000)

#generate Q-Q plot for the dataset
fig = sm.qqplot(data, line='45')
plt.show()
```
