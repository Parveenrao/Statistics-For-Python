```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample1 = np.random.normal(size = 300)
sample2 = np.random.uniform(size = 300)
```

```python
plt.hist(sample1 , density = True )
plt.title('Normal Distribution')
sns.kdeplot(sample1 , color = 'green')
plt.show()
```

```python
plt.hist(sample2 , density = True  )
plt.title('Normal Distribution')
sns.kdeplot(sample2 , color = 'red')
plt.show()
```

```python
from scipy.stats import shapiro
```

```python
stat , p_value = shapiro(sample1)

print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"P-value: {p_value}")
```

```python
if p_value > 0.05:
    print("The data is normally distributed (fail to reject H).")
else:
    print("The data is not normally distributed (reject H).")
```

```python
stat , p_value = shapiro(sample2)

print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"P-value: {p_value}")
```

```python
if p_value > 0.05:
    print("The data is normally distributed (fail to reject H).")
else:
    print("The data is not normally distributed (reject H).")
```
