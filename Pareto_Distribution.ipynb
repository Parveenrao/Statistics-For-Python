```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scipy
```

```python
from scipy.stats import pareto
```

```python
alpha_param = 2
sample_size = 1000

data = pareto.rvs(alpha_param , size = sample_size)
plt.hist(data , density = True , bins = 30 ,  edgecolor = 'black' , color = 'blue')
sns.kdeplot(data , color = 'red')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title('Pareto Distribution')

plt.show()
```
