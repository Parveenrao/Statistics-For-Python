```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# mean of the distribution
mean = 100
#standard Deviation of the distribution 
standard_deviation = 5
# size
size = 1000
```

```python
values = np.random.normal(mean , standard_deviation , size)
```

```python
plt.hist(values, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')
sns.kdeplot(values, color='green', linewidth=2)
plt.show()
```
